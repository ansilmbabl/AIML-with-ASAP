#### 6. boxplot
used to plot quartiles (%)

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df
```

```python
df.info() # different properties of the columns
```

plotting boxplot

```python
sns.boxplot(data = df, x = 'age')
```

```python
# OR
sns.boxplot(x = df['age'])
```

```python
df['age'].value_counts()
```

```python
df['age'].min()
```

```python
df['age'].median()
```

```python
df['age'].max()
```

```python
df['age'].quantile(.25)
```

```python
df['age'].quantile(.75)
```

```python
df['age'].quantile(1)
```

```python
sns.boxplot(data = df, x = 'age', y = 'class')

# age = numerical varible
# class = categorical variable
```

```python
sns.boxplot(data = df, x = 'age', y = 'class', hue = 'alone')
```

```python
sns.boxplot(data = df, x = 'age', y = 'class', hue = 'alone', order = ['Third','Second','First'])
```

```python
sns.boxplot(data = df, x = 'age', y = 'class', hue = 'alone', hue_order = [1,0])
```

```python
sns.boxplot(data = df, x = 'age', y = 'class', hue = 'alone', hue_order = [0,1], notch = True)
```

```python
sns.boxplot(data = df, x = 'age', y = 'class', hue = 'alone', hue_order = [False,True], notch = True)
```

```python
sns.boxplot(data = df, x = 'age', y = 'class', hue = 'alone', hue_order = [0,1], showcaps = False)
```

```python
sns.boxplot(data = df, x = 'age')
```

```python
sns.boxplot(data = df, x = 'age', boxprops = {'facecolor':'g'})
```

```python
sns.boxplot(data = df, x = 'age', boxprops = {'facecolor':'g'}, medianprops = {'color':'red'})
```

```python
sns.boxplot(data = df, x = 'age', boxprops = {'facecolor':'g'}, medianprops = {'color':'red'}, notch = True)
```

```python
sns.boxplot(data = df, x = 'age', boxprops = {'facecolor':'g'}, medianprops = {'color':'red'}, notch = True, showcaps = False)
```

#### 7. Kernel Density Estimation (KDE)

```python
tips = sns.load_dataset('tips')
sns.kdeplot(data = tips, x = 'total_bill')
```

```python
tips
```

```python
sns.kdeplot(data=tips)
```

```python
sns.kdeplot(data = tips, x = 'total_bill', hue = 'time')
```

```python
sns.kdeplot(data = tips, x = 'total_bill', hue = 'time', multiple = 'stack')
```

```python
sns.kdeplot(data = tips, x = 'total_bill', hue = 'time', multiple = 'layer')
```

```python
sns.kdeplot(data = tips, x = 'total_bill', hue = 'time', multiple = 'fill')
```

#### 8.Violin plot
combination of boxlot and KDE   
gives summary as well as distribution

```python
df = sns.load_dataset('titanic')
sns.violinplot(data = df)
```

```python
sns.violinplot(data = df, x = 'age')
```

```python
sns.violinplot(data = df, x = 'age', y = 'class')
```

```python
sns.violinplot(data = df, x = 'age', y = 'class', hue = 'alone')
```

```python
sns.violinplot(data = df, x = 'age', y = 'class', hue = 'alone', hue_order=[True,False])
```

```python
sns.violinplot(data = df, y = 'age', x = 'class', hue = 'alone', hue_order=[True,False])
```

representing every distribution

```python
sns.violinplot(data = df, x = 'age', y = 'class', inner = 'stick')
```

```python
sns.violinplot(data = df, x = 'age', y = 'class', inner = 'box')
```

```python
sns.violinplot(data = df, x = 'age', y = 'class', inner = 'quartile')
```

```python
sns.violinplot(data = df, x = 'age', y = 'class', inner = None)
```

```python
sns.violinplot(data = df, x = 'age', y = 'class', inner = 'point')
```
