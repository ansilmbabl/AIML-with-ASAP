```python
cars = sns.load_dataset('mpg')
cars
```

```python
cars.groupby('origin')
```

```python
cars.groupby('origin').cylinders
```

```python
cars.groupby('origin').cylinders.value_counts()
```

this is not in the form of a DataFrame  
unstaking it 

```python
cars.groupby('origin').cylinders.value_counts().unstack()
```

filling NaN value by 0

```python
cylinder = cars.groupby('origin').cylinders.value_counts().unstack().fillna(0)
cylinder
```

```python
sns.heatmap(cylinder)
```

taking trnspose of the data

```python
cylinder.T
```

```python
sns.heatmap(cylinder.T)
```

correlation heatmap

```python
cars.corr()
```

```python
sns.heatmap(cars.corr())
```

changing the color transition 

```python
sns.heatmap(cars.corr(),cmap = 'crest')
```

```python
sns.heatmap(cars.corr(),cmap = 'Blues')
```

changing color to both extend from centre

```python
sns.heatmap(cars.corr(),cmap = 'RdBu')
```

setting centre to zero

```python
sns.heatmap(cars.corr(),cmap = 'RdBu', center = 0)
```

setting min and max valu for colorbar

```python
sns.heatmap(cars.corr(),cmap = 'RdBu', center = 0, vmin = -1, vmax= 1) 
```

annotating the color values

```python
cylinder
```

```python
sns.heatmap(cylinder, cmap= 'Blues', annot= True)
```

cganing scientific notation

```python
sns.heatmap(cylinder, cmap= 'Blues', annot= True, fmt = '.0f')
```

```python
sns.heatmap(cylinder, cmap= 'Blues', annot= True, fmt = '.1f')
```

setting different properties of annotation


```python
sns.heatmap(cylinder, cmap= 'Blues', annot= True, fmt = '.0f', annot_kws= {'fontsize':18})
```

```python
sns.heatmap(cylinder, cmap= 'Blues', annot= True, fmt = '.0f', annot_kws= {'fontsize':18,'fontweight':'bold','fontfamily':'serif'})
```

```python
sns.heatmap(cylinder, cmap= 'Blues', annot= True, fmt = '.0f', annot_kws= {'fontsize':18,'fontweight':'bold','fontfamily':'serif'},linewidth = 2, linecolor= 'g')
```

```python
sns.heatmap(cars.corr(),cmap='RdBu',linecolor='black',linewidths=5)
```

#### 10. jointplot
used for plotting association and distribution with two varibale  
combination of $scatterplot$ and $histogram$


```python
geyser = sns.load_dataset('geyser')
geyser
```

```python
sns.jointplot(geyser)
```

```python
sns.jointplot(x = 'waiting', y = 'duration', data = geyser)
```

changing the kind

```python
sns.jointplot(x='waiting', y = 'duration', data = geyser, kind = 'scatter')
```

```python
sns.jointplot(x='waiting', y = 'duration', data = geyser, kind = 'kde')
```

```python
sns.jointplot(x='waiting', y = 'duration', data = geyser, kind = 'hex')
```

```python
sns.jointplot(x='waiting', y = 'duration', data = geyser, kind = 'reg')
```

```python
sns.jointplot(x='waiting', y = 'duration', data = geyser, kind = 'resid')
```

```python
sns.jointplot(x='waiting', y = 'duration', data = geyser, kind = 'hist')
```

using a categorical variable by hue

```python
sns.jointplot(x='waiting', y = 'duration', data = geyser, hue = 'kind')
```

```python
sns.jointplot(x='waiting', y = 'duration', data = geyser, hue = 'kind', kind = 'kde')
```
