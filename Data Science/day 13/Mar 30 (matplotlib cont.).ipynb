# Matplotlib cont.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
```

```python
# set of subplots using Object oreinted approach

fig,axes = plt.subplots(nrows=1 ,ncols=3)
x = np.linspace(0,5,11)

axes[0].plot(x,x)
axes[0].set_title('linear')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y = x')

axes[1].plot(x,x**2)
axes[1].set_title('quadratic')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y = x^2')

axes[2].plot(x,x**3)
axes[2].set_title('cubic')
axes[2].set_xlabel('x')
axes[2].set_ylabel('y = x^3')

fig.tight_layout()
```

#### 1. plot(x,y)

```python
x = np.linspace(0,10,500)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
# changing style
plt.style.use('dark_background')
x = np.linspace(0,10,500)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
# changing style
plt.style.use('bmh')
x = np.linspace(0,10,500)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x,y)
plt.grid() # with grid
plt.show()
```

```python
# changing style
plt.style.use('default')
x = np.linspace(0,10,500)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x,y)
plt.grid()
plt.show()
```

## 3D scatterplot

```python
x = np.random.uniform(23,32,50)
y = np.random.uniform(0,100,50)
z = np.random.uniform(-30,-15,50)

fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})
ax.scatter(x,y,z)
plt.show()
```

```python
# 4D (https://stackoverflow.com/questions/14995610/how-to-make-a-4d-plot-with-matplotlib-using-arbitrary-data)

x = np.random.uniform(23,32,50)
y = np.random.uniform(0,100,50)
z = np.random.uniform(-30,-15,50)
c = np.random.uniform(-30,-15,50)

fig,ax = plt.subplots(subplot_kw = {'projection':'3d'})
ax.scatter(x,y,z,c=c, cmap=plt.hot())
# fig.colorbar(img)
plt.show()
```
