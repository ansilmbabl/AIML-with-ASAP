```python
# creating subplots
x = np.linspace(0,5,11)

plt.subplot(2,2,1)
plt.plot(x,x,'go--') #fmt parmater
plt.xlabel('x')
plt.ylabel('y = x')
plt.annotate('(2,4)',xy=(2,4))
plt.title('linear')

plt.subplot(2,2,2)
plt.plot(x,x**2,'r*--')
plt.xlabel('x')
plt.ylabel('Y = x^2')
plt.annotate('(2,4)',xy=(2,4))
plt.title('quadratic')

plt.subplot(2,2,3)
plt.plot(x,x**3,'bs:')
plt.xlabel('x')
plt.ylabel('y = x^3')
plt.title('cubic')

plt.subplot(2,2,4)
plt.plot(x,x**2,'m^:')
plt.xlabel('x')
plt.ylabel('Y = log(x)')
plt.title('logarithmic')

plt.tight_layout()
```

### 2. Object oriented approach in matplotlib
create figure objects  
each objects will have different methods and attributes

```python
fig = plt.figure(facecolor='g',figsize=(10,8)) # creating figure canvas
```

```python
# Set up matplotlib to work interactively.
# %matplotlib inline 

# creating axes (canvas) - area under which fig is drawing
x = np.array([1,3,4,6,8,10]) 
y = np.array([6,8,4,1,3,10])
axes = fig.add_axes([0.1,0.8,0.5,0.5])
axes.set_xlabel('x values')
axes.set_ylabel('y values')
axes.plot(x,y,'b')
fig
```

```python
ax2 = fig.add_axes([.2,0.5,0.4,0.3])
ax2.set_xlabel('inner x')
ax2.set_ylabel('inner y')
ax2.set_title('inner plot')
ax2.plot(x,x+y)
fig
```

```python
# set of aubplots using Object oreinted approach

fig,axes = plt.subplots(nrows=1 ,ncols=3)
x = np.linspace(0,5,11)

axes[0].plot(x,x)
axes[0].set_title('linear')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y = x')

axes[1].plot(x,x**2)
axes[1].set_title('quadratic')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y = x^2')

axes[2].plot(x,x**3)
axes[2].set_title('cubic')
axes[2].set_xlabel('x')
axes[2].set_ylabel('y = x^3')

fig.tight_layout()
```

```python
# same question with for loop
fig,axes = plt.subplots(nrows=1 ,ncols=3)
x = np.linspace(0,5,11)

for i in range(len(axes)):
    axes[i].plot(x,x**(i+1))
    
print(type(fig),type(axes))
```
