```python
# importing library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

plotting with some basic data

```python
x = [10,20,30,40]
y = [20,25,35,55]
```

```python
#plotting values
plt.plot(x,y)

plt.title('Linear graph',fontsize = 10 ,color = 'green') #title to the graph

#labels (x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

#limits (x,y) - scales
plt.xlim(0,60)
plt.ylim(0,80)

plt.legend(['abcd','efgh']) #legends
plt.show() #showing the graph
```

### 1. Functional method in matplotlib

```python
x = np.linspace(0,5,11)
x
```

```python
plt.plot(x,x)
plt.show()
```

```python
plt.plot(x,x**2)
plt.show()
```

```python
plt.plot(x,x**3)
plt.show()
```

```python
plt.plot(x,np.log(x))
plt.show()
```

```python
plt.title('Graph',color = 'red',fontsize = 15)
plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3,label='cubic')
plt.plot(x,np.log(x),label='logarithmic')

plt.legend()

plt.xlim(0,7)
plt.ylim(0,150)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```
