#### 1.line plot

plotting in matplotlib

```python
l1 = [1,2,3,4,5]
l2 = [1,4,9,16,25]
```

```python
plt.plot(l1,l2)
plt.show()
```

same thing using seaborn

```python
sns.lineplot(x = l1, y = l2)
plt.show()
```

using data frames 

```python
df = pd.DataFrame({'x_values':l1,'y_values':l2})
df
```

```python
sns.lineplot(x = 'x_values', y = 'y_values', data = df)
plt.show()
```

#### 2. bar plot

```python
df = sns.load_dataset('penguins')
df
```

```python
sns.barplot(x = 'island', y = 'body_mass_g', data = df)
plt.show()
```

x and y should be numeric

```python
sns.barplot(x = 'island', y = 'sex', data = df)
plt.show()
```

differnertiationg bars based on another property

```python
sns.barplot(x = 'island', y = 'body_mass_g', data = df, hue='sex')
plt.show()
```

changing order

```python
sns.barplot(x = 'island', y = 'body_mass_g', data = df, hue='sex', hue_order = ("Female",'Male'))
plt.show()
```

```python
sns.barplot(x = 'island', y = 'body_mass_g', data = df, hue='sex', hue_order = ("Female",'Male'), order = ('Dream','Torgersen'))
plt.show()
```

```python
sns.barplot(x = 'island', y = 'body_mass_g', data = df, hue='sex', hue_order = ("Female",'Male'), order = ('Dream','Torgersen','Biscoe'))
plt.show()
```

to remove confident interval

```python
sns.barplot(x = 'island', y = 'body_mass_g', data = df, hue='sex', hue_order = ("Female",'Male'), order = ('Dream','Torgersen','Biscoe'), ci = 0)
plt.show()
```

```python
sns.barplot(x = 'island', y = 'body_mass_g', data = df, hue='sex', hue_order = ("Female",'Male'), order = ('Dream','Torgersen','Biscoe'), ci = 0, estimator = np.mean)
plt.show()
```

#### 3. histplot()

```python
df
```

```python
sns.histplot(data = df, x='flipper_length_mm')
plt.show()
```

```python
sns.histplot(data = df, y='flipper_length_mm')
plt.show()
```

specifying number of bins in between

```python
sns.histplot(data = df, x='flipper_length_mm',binwidth=3)
plt.show()
```

specifying total number of bins

```python
sns.histplot(data = df, x='flipper_length_mm',bins=50)
plt.show()
```

KDE kernal density plot

```python
sns.histplot(data = df, x = 'flipper_length_mm', kde=True)
plt.show()
```

```python
sns.histplot(data = df, x='flipper_length_mm',hue='species')
plt.show()
```

stacking them


```python
sns.histplot(data = df, x='flipper_length_mm',hue='species', multiple='stack')
plt.show()
```

stepping

```python
sns.histplot(data = df, x='flipper_length_mm',hue='species',element="step")
plt.show()
```

datapoints connecting using polygons

```python
sns.histplot(data = df, x='flipper_length_mm',hue='species',element="poly")
plt.show()
```

```python
f,ax = plt.subplots(2,2,figsize = (15,15))
m=['layer','dodge','stack','fill']
for i,j in zip(ax.flat,m):
    sns.histplot(data = df, x='flipper_length_mm',hue='species',multiple=j,ax=i)
plt.show()
```

```python
f,ax = plt.subplots(2,2,figsize = (15,15))
m=['layer','dodge','stack','fill']
for i,j in ax,m:
    sns.histplot(ax=i,data = df, x='flipper_length_mm',hue='species',multiple=j)
plt.show()
```

changing y axis value from count 

```python
sns.histplot(data=df, x='flipper_length_mm',stat = 'probability')
```

```python
sns.histplot(data=df, x='flipper_length_mm',stat = 'percent')
```

#### 4. relplot (relationship plot)
scatterplot and lineplots are two variation of relationship plot

```python
tips = sns.load_dataset('tips')
tips
```

plotting data

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'line')
```

differentiating based on another value (day), this is not possible in normal scatter plot

```python
tips.day.unique()
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', kind='line')
```

creating subplots across categpry which is not possible in normal scatter plot

```python
tips.time.unique()
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', col='time')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', col='time', kind = 'line')
```

changing with rows also  
row = sex  
column = time

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', col='time', row = 'sex')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', col='time', row = 'sex', kind = 'line')
```

category with days

```python
sns.relplot(x='total_bill', y = 'tip', data = tips , hue = 'time', col = 'day')
```

wrapping column

```python
sns.relplot(x='total_bill', y = 'tip', data = tips , hue = 'time', col = 'day',col_wrap = 2)
```

changing marker style

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', col='time')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', col='time', style = 'sex')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', col='time', markers = True, style = 'sex')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', col='time', markers = ['<','>'], style = 'sex')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', col='time', markers = True, style = 'size')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', col='time', markers = ['<','>'], style = 'sex',size = 'size')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', col='time', markers = ['<','>'], style = 'sex',size = 'tip')
```
