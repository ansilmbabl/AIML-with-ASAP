# Outliers

```python
import pandas as pd
l=[11,10,12,14,18,19,17,11,10,18,0,17,16,111,11,12,198,72,112,16,10,19,1]
```

converting to dataframe

```python
df = pd.DataFrame(l,columns=['Data'])
df
```

sorting the value

```python
df =df.sort_values(by='Data')
df
```

finding quantile

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3 - q1
iqr
```

lower & upper bounds

```python
lb = q1 - (1.5 * iqr)
ub = q3 + (1.5 * iqr)
print(lb,ub)
```

```python
df[df<lb]
```

```python
df[df>ub]
```

```python
df_woo = df[(df<=ub) & (df>=lb) ]
df_woo
```
