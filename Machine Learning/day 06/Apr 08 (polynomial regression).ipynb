# Polynomial Regression
Linear regression problem in nonlinear data

```python
import numpy as np
x = np.array([2,3,4,5,6,7,7,8,9,11,12])
y = np.array([18,16,15,17,20,23,25,28,31,30,29])
```

```python
print(len(x),len(y))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.3, random_state=0)
X_train
```

```python
X_train,X_test = X_train.reshape(-1,1),X_test.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,Y_train)
```

```python
plt.plot(x,y,"b")
```

```python
plt.plot(X_train,reg.predict(X_train),'r')
```

```python
plt.scatter(x,y)
plt.plot(X_train,reg.predict(X_train),'r')
```

#### plynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 2)
X_train_trans = poly.fit_transform(X_train)
X_test_trans = poly.fit_transform(X_test)
reg = LinearRegression()
reg.fit(X_train_trans,Y_train)

y_pred = reg.predict(X_train_trans)
y_pred
```

```python
X_train
```

```python
X_train_trans
```

```python
X_grid_train = np.arange(min(X_train),max(X_train)+0.1, step = 0.01)
X_grid_test = np.arange(min(X_test),max(X_test)+0.1, step = 0.01)

X_grid_train = X_grid_train.reshape(-1,1)
X_grid_test = X_grid_test.reshape(-1,1)
```

```python
plt.plot(x,y,'b')
plt.plot(X_grid_train,reg.predict(poly.transform(X_grid_train)),'g')
plt.show()
```
