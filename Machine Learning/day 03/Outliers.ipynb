# Outliers

```python
import pandas as pd
l=[11,10,12,14,18,19,17,11,10,18,0,17,16,111,11,12,198,72,112,16,10,19,1]
```

converting to dataframe

```python
df = pd.DataFrame(l,columns=['Data'])
df
```

sorting the value

```python
df =df.sort_values(by='Data')
df
```

finding quantile

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3 - q1
iqr
```

lower & upper bounds

```python
lb = q1 - (1.5 * iqr)
ub = q3 + (1.5 * iqr)
print(lb,ub)
```

```python
df[df<lb]
```

```python
df[df>ub]
```

```python
df_woo = df[(df<=ub) & (df>=lb) ]
df_woo
```

```python
df.describe()
```

```python
df_woo.describe()
```

## SD method 

```python
import pandas as pd
l=[11,10,12,14,18,19,17,11,10,18,0,17,16,111,11,12,198,72,112,16,10,19,1]
```

```python
df = pd.DataFrame(l,columns=['Data'])
df
```

```python
df =df.sort_values(by='Data')
df
```

```python
# upper and lower bound

ub = df.Data.mean() + 3 * df.Data.std()
lb = df.Data.mean() - 3 * df.Data.std()

print(ub,lb)
```

```python
df_woo = df[(df.Data < ub) & (df.Data > lb)]
df_woo
```

## Z score method

```python
import pandas as pd
l=[-189,-200,11,10,12,14,18,19,17,11,10,18,0,17,16,111,11,12,198,72,112,16,10,19,1]
df = pd.DataFrame(l,columns=['Data'])
df
```

```python
df =df.sort_values(by='Data')
df
```

```python
df['Zscore'] = (df.Data - df.Data.mean()) / df.Data.std()
df
```

```python
df[df['Zscore']>2.5]
```

```python
df[df['Zscore']<-2.5]
```

```python
df_woo = df[(df['Zscore']<2.5) & (df['Zscore']>-2.5)]
df_woo
```
