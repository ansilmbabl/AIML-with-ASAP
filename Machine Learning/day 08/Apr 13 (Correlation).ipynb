# Spearman and Pearson Correlation

```python
df = pd.read_csv("D:\AIML\ML\data\wine_dataset.csv")
df.head()
```

```python
from scipy.stats import spearmanr
stat, p = spearmanr(df[:]['volatile_acidity'],df[:]['residual_sugar'])
print(f'p = {p:0.9f}')
print(p,stat)
```

```python
from scipy.stats import pearsonr
stat, p = pearsonr(df[:]['volatile_acidity'],df[:]['residual_sugar'])
print(f'p = {p:0.9f}')
print(p,stat)
```
